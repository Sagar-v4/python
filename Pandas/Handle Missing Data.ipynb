```python
import pandas as pd
```

## Handalling missing data

1. isnull()
2. notnull()
3. dropna()
4. fillna()
5. replace()
6. interpolate()

```python
df = pd.read_csv("companies_missing.csv")

df.head()
```

### isnull()

```python
# Field is null?

df = pd.read_csv("companies_missing.csv")

df.isnull()
```

```python
# rows where cell is null in specific col

df = pd.read_csv("companies_missing.csv")

data = pd.isnull(df["Criteria"])

df[data]
```

### nutnull()

```python
# field is not null?

df = pd.read_csv("companies_missing.csv")

df.notnull()
```

```python
# rows where cell is not null in specific col

df = pd.read_csv("companies_missing.csv")

data = pd.notnull(df["Criteria"])

df[data]
```

### dropna()

```python
# null value rows will be drop | axis = 0 (default)

df = pd.read_csv("companies_missing.csv")

df.dropna()
```

```python
# null value rows will be drop | axis = 1

df = pd.read_csv("companies_missing.csv")

df.dropna(axis = 1)
```

```python
# drop all rows where any cell is null

df = pd.read_csv("companies_missing.csv")

df.dropna(how = "any")
```

```python
# drop all rows where all cell is null

df = pd.read_csv("companies_missing.csv")

df.dropna(how = "all")
```

```python
# drop all rows where specific columns cell is null

df = pd.read_csv("companies_missing.csv")

df.dropna(subset = ["Criteria", "Type"])
```

```python
# replaces DataFrame by removing all rows where any cell is null

df = pd.read_csv("companies_missing.csv")

df.dropna(inplace = True)

df
```

```python
# drops all the row where null value cell count of row == thresh

df = pd.read_csv("companies_missing.csv")

df.dropna(thresh = 3)
```

### fillna()

```python
# fill all null cells with specific given value

df = pd.read_csv("companies_missing.csv")

df.fillna("---")
```

```python
# fill specific value per columns by giving dictionary

data = {
    "Company Name": "MY COMPANY NAME",
    "Type": "TIME PASS",
    "Category": "AAAA",
    "Criteria": 100.0,
    "Placement Date": "ANYTHING",
    "Open For": "ONLY MSC IT"
} # dict

df = pd.read_csv("companies_missing.csv")

df.fillna(data)
```

```python
# fill the null cells with previous / upper values = forward filling

df = pd.read_csv("companies_missing.csv")

df.fillna(method = "ffill")
```

```python
# fill the null cells with next / lower values = backward filling

df = pd.read_csv("companies_missing.csv")

df.fillna(method = "bfill")
```

```python
# fill the null cells with next / lower values = backward filling with axis = 1

df = pd.read_csv("companies_missing.csv")

df.fillna(method = "ffill", axis = 1)
```

```python
# replaces DataFrame by adding specific value to all null cells

df = pd.read_csv("companies_missing.csv")

df.fillna("----", inplace = True)

df
```

```python
# fill columns null cell by limits (only fist limit number of cell per column)
# Examle:
# col -> updated limit = 1 - > col -> val = "something"
# ---                          ---
#                              "something"
# "fd"                         "fd"
#
# "sd"                         "sd"

df = pd.read_csv("companies_missing.csv")

df.fillna("----", limit = 2)
```

### replace()

```python
# replaces specific values

df = pd.read_csv("companies_missing.csv")

df.replace(to_replace = "A1", value = "AAAAAA")
```

```python
# replaces specific values

df = pd.read_csv("companies_missing.csv")

dataArr = ["job", "Unisys", "A1", ]

df.replace(dataArr, "-----0000----")
```

```python
# replaces specific values with regex

df = pd.read_csv("companies_missing.csv")

regex = "[A-Za-z]"

df.replace(regex, "-----0000----", regex = True)
```

```python
# replaces specific values with dictionary & regex

df = pd.read_csv("companies_missing.csv")

regex = {
    "Category": "[A-z0-9]"
} # dictionary

df.replace(regex, "-----0000----", regex = True)
```

```python
# replaces specific values with forward filling

df = pd.read_csv("companies_missing.csv")

df.replace("A1", method = "ffill")
```

```python
# replaces specific values with backward filling

df = pd.read_csv("companies_missing.csv")

df.replace("A", method = "bfill")
```

```python
# replaces specific values with limit data

df = pd.read_csv("companies_missing.csv")

df.replace("A", method = "bfill", limit = 3)
```

```python
# replaces specific values in original DataFrame

df = pd.read_csv("companies_missing.csv")

df.replace("A", method = "bfill", limit = 3, inplace = True)

df
```

### interpolate()

```python
# automatically fill null data by its previous sequence observations
# example: [1, 2, 3, 4, null, 6, 7, 8, 8] -> null will filled by 5 value automatic
# ONLY WORKS WITH NUMBERS

df = pd.read_csv("companies_missing.csv")

df.interpolate()
```

```python
# automatically fill null data by its previous sequence observations
# example: [1, 2, 3, 4, null, 6, 7, 8, 8] -> null will filled by 5 value automatic
# ONLY WORKS WITH NUMBERS

df = pd.read_csv("companies_missing.csv")

df.interpolate()
```
