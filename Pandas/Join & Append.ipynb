# Join

```python
data1 = {
    "A": [1, 2, 3],
    "B": [9, 8, 7]
} 

data2 = {
    "Y": [11, 22, 33],
    "Z": [99, 88, 77]
} 

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

df1.join(df2)
```

```python
data1 = {
    "A": [1, 2, 3, 4, 5],
    "B": [9, 8, 7, 6, 5]
} 

data2 = {
    "Y": [11, 22],
    "Z": [99, 88]
} 

indexes = ["a", "b", "c", "d", "e", "f"]

df1 = pd.DataFrame(data1, index = indexes[: len(data1["A"])])
df2 = pd.DataFrame(data2, index = indexes[: len(data2["Z"])])

df2.join(df1, how = "right") # join - left, right, inner, outer
```

```python
data1 = {
    "A": [1, 2, 3, 4, 5],
    "B": [9, 8, 7, 6, 5]
} 

data2 = {
    "Y": [11, 22],
    "B": [99, 88]
} 

indexes = ["a", "b", "c", "d", "e", "f"]

df1 = pd.DataFrame(data1, index = indexes[: len(data1["A"])])
df2 = pd.DataFrame(data2, index = indexes[: len(data2["Y"])])

df2.join(df1, how = "outer", lsuffix = "_lsuffix", rsuffix = "_rsuffix")
```

# Append - depricated

```python
data1 = {
    "A": [1, 2, 3, 4, 5],
    "B": [9, 8, 7, 6, 5]
} 

data2 = {
    "Y": [11, 22],
    "B": [99, 88]
} 

indexes = ["a", "b", "c", "d", "e", "f"]

df1 = pd.DataFrame(data1, index = indexes[: len(data1["A"])])
df2 = pd.DataFrame(data2, index = indexes[: len(data2["Y"])])

df1.append(df2)
# pd.concat([df1, df2])
```

```python
data1 = {
    "A": [1, 2, 3, 4, 5],
    "B": [9, 8, 7, 6, 5]
} 

data2 = {
    "Y": [11, 22],
    "B": [99, 88]
} 

indexes = ["a", "b", "c", "d", "e", "f"]

df1 = pd.DataFrame(data1, index = indexes[: len(data1["A"])])
df2 = pd.DataFrame(data2, index = indexes[: len(data2["Y"])])

df1.append(df2, ignore_index = True)
# pd.concat([df1, df2])
```
