```python
import pandas as pd
```

```python
com = pd.read_csv("companies.csv")

com
```

```python
# Specific number of rows

com = pd.read_csv("companies.csv", nrows = 3)

com
```

```python
# Specific cols

com = pd.read_csv("companies.csv", nrows = 3, usecols = ["Company Name", "Open For"])

com
```

```python
# Specific cols with their indexes

com = pd.read_csv("companies.csv", nrows = 3, usecols = [0, 5])

com
```

```python
# Skip rows by row numbers

com = pd.read_csv("companies.csv", nrows = 3, usecols = [0, 5], skiprows = [0, 1])

com
```

```python
# Index call

com = pd.read_csv("companies.csv", nrows = 3, index_col = "Company Name")

com
```

```python
# Headers

com = pd.read_csv("companies.csv", nrows = 3, header = 2)

com
```

```python
# Headers

com = pd.read_csv("companies.csv", nrows = 3, names = ["A", "B", "C", "D", "E", "F"])

com
```

```python
# Headers

com = pd.read_csv("companies.csv", nrows = 3, header = None)

com
```

```python
# Change column data type

types = {
    "Criteria": "float"
}
com = pd.read_csv("companies.csv", nrows = 2, skiprows = [1], dtype = types)

com
```
