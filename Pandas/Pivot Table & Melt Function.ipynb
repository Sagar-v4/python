```python
import pandas as pd
```

#### Used to reshape file

# melt()

```python
data = {
    "Role_No": [1, 2, 3, 4],
    "Eng": [34, 67, 76, 23],
    "Sci": [65, 76, 34, 77],
} 

df = pd.DataFrame(data)

df
```

```python
pd.melt(df)
```

```python
ids = ["Role_No"]

pd.melt(df, id_vars = ids)
```

```python
ids = ["Role_No"]
varName = "Subject"
valueName = "Marks"

pd.melt(df, id_vars = ids, var_name = varName, value_name = valueName)
```

# privot()

```python
data = {
    "Role_No": [1, 2, 3, 4],
    "ids": ["a", "b", "c", "d"],
    "Eng": [34, 67, 76, 23],
    "Sci": [65, 76, 34, 77],
} 

df = pd.DataFrame(data)

df
```

```python
df.pivot(index = "Role_No", columns = "ids")
```

```python
df.pivot_table(index = "ids", columns = "Role_No", aggfunc = "mean")
```

```python
df.pivot_table(index = "ids", columns = "Role_No", aggfunc = "sum", margins = "True")
```
