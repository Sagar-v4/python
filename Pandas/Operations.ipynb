```python
import pandas as pd
```

```python
# data = [1, 2, 3, 4] # 1D list
# data = [[1, 2], [3, 4]] # 2D list
data = {"a": [1, 2, 3], "b": [4, 5, 6]} # dictionary

df = pd.DataFrame(data)

df
```

```python
# Addition

df["Addition"] = df["a"] + df["b"]

df
```

```python
# Subtraction

df["Subtraction"] = df["a"] - df["b"]

df
```

```python
# Multipy

df["Multipy"] = df["a"] * df["b"]

df
```

```python
# Divide

df["Divide"] = df["a"] / df["b"]

df
```

```python
# Specific data

data = {"a": [4, 6, 2, 1, 7, 4, 2], "b": [7, 3, 6, 8, 5, 2, 1]} # dictionary

df = pd.DataFrame(data)

df["a_evens"] = df["a"] % 2 == 0
df["b_evens"] = df["b"] % 2 == 0

df
```
