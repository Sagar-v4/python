```python
import pandas as pd
```

# Insert(index, name, newValue)

```python
data = {
    "i": [1, 2, 3, 4, 5, 6, 7],
    "ii": [4, 5, 6, 7, 8, 9, 10],
}

df = pd.DataFrame(data)

df
```

```python
# insert

ind = 2
name = "iii"
value = [9, 8, 7, 6, 5, 4, 3]
df.insert(ind, name, value)

df
```

```python
# copy with slicing - specific length

df["i_copy"] = df["ii"][1:3]
df["ii_copy"] = df["ii"][6:3:-1]
df["iii_copy"] = df["iii"]

df
```

# del | pop()

```python
# pop()

deleted = df.pop("i")

print(f"Deleted: \n{deleted}")
print(f"Original: \n{df}")
```

```python
# del

del df["ii"]

print(f"Original: \n{df}")
```
