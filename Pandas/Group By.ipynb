# Group By

```python
data = {
    "Names": ["d", "g", "e", "s", "d","g", "e", "s", "y", "q", "y", "q", "a"],
    "Sub_1": [11, 22, 33, 44, 11, 44, 65, 2, 33, 22, 11, 45, 66],
    "Sub_2": [31, 52, 36, 23, 76, 1, 54, 72, 74, 23, 54, 25, 46],
}

df = pd.DataFrame(data)
```

```python
names = df.groupby("Names")

print("\nindex - name - sub1 - sub2")
for key, value in names:
    print()
    print(key)
    print(value)
```
