```python
import pandas as pd
```

### Used to combine 2 DataFrames

# merge()

```python
data1 = {
    "A": [1, 2, 3, 4],
    "B": [11, 22, 33, 44]
}

data2 = {
    "A": [1, 2, 3, 4],
    "C": [111, 222, 333, 444]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

pd.merge(df1, df2)
```

```python
data1 = {
    "A": [1, 2, 3, 4],
    "B": [11, 22, 33, 44]
}

data2 = {
    "A": [1, 20, 3, 40],
    "C": [111, 222, 333, 444]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

pd.merge(df1, df2)
```

```python
data1 = {
    "A": [1, 2, 3, 4],
    "B": [11, 22, 33, 44]
}

data2 = {
    "A": [1, 20, 3, 40],
    "C": [111, 222, 333, 444]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

pd.merge(df1, df2, how = "left") # how (join) - left, right, inner, outer
```

```python
data1 = {
    "A": [1, 2, 3, 4],
    "B": [11, 22, 33, 44]
}

data2 = {
    "A": [1, 20, 3, 40],
    "C": [111, 222, 333, 444]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

pd.merge(df1, df2, how = "outer", indicator = True) # how (join) - left, right, inner, outer
```

```python
data1 = {
    "A": [1, 2, 3, 4],
    "B": [11, 22, 33, 44]
}

data2 = {
    "A": [1, 20, 3, 40],
    "B": [111, 222, 333, 444]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

pd.merge(df1, df2, left_index = True, right_index = True)
```

```python
data1 = {
    "A": [1, 2, 3, 4],
    "B": [11, 22, 33, 44]
}

data2 = {
    "A": [1, 20, 3, 40],
    "B": [111, 222, 333, 444]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

pd.merge(df1, df2, left_index = True, right_index = True, suffixes = ["_ID", "_Name"])
```

# concat()

```python
data1 = [1, 2, 3, 4]
data2 = [111, 222, 333, 444]

df1 = pd.Series(data1)
df2 = pd.Series(data2)

pd.concat([df1, df2])
```
